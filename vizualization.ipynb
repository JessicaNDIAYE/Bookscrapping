```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
# Chargement des fichiersimport pandas as pd
top_authors = pd.read_csv("top_authors.csv", index_col=0)
avg_rating = pd.read_csv("avg_rating_by_genre.csv", index_col=0)
```

```python


plt.figure(figsize=(10,6))
sns.barplot(x=top_authors.values.flatten(), y=top_authors.index, palette="mako")
plt.title("Top 10 auteurs les plus présents dans les bestsellers Amazon")
plt.xlabel("Nombre de livres")
plt.ylabel("Auteur")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x=avg_rating.values.flatten(), y=avg_rating.index, palette="coolwarm")
plt.title("Note moyenne par genre")
plt.xlabel("Note moyenne")
plt.ylabel("Genre")
plt.xlim(4.5, 4.7)  # pour zoomer sur les valeurs proches
plt.tight_layout()
plt.show()
```
